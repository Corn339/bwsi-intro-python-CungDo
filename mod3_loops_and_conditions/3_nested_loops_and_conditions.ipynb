## FizzBuzz Problem

Write a short program that prints each number from 1 to 100 except:

- For each multiple of 3, print "Fizz" instead of the number

- For each multiple of 5, print "Buzz" instead of the number

- For numbers which are multiples of both 3 and 5, print "FizzBuzz" instead of the number

### Q.1 Please outline in words the steps that should be executed for each iteration of the loop

### A.1 YOUR ANSWER HERE
Essentially, I'm gonna create an integer variable equaling 0, and while the number is less than or equal to 100, it'll print it out, and inside I'l include an if statement asking if the number is divisible by 3 or 5, thus printing fizzbuzz, ending with an else statement to just print the number if these conditions are not met.

### Q.2 Please implement the solution described in Q.1 with comments

```python
current_num = 1
while current_num <=100:
    if current_num%3==0 or current_num%5==0:
        print("FizzBuzz")
    else:
        print(current_num)
    current_num+=1
```

### Q.3 Please complete the commented lines below with descriptions of the code as indicated. 

```python
# Create list to store prime numbers
prime_numbers_list = []

# This for loop iterates over the range of numbers to check for primality
for n in range(2,11):
    # Set is_prime as True
    is_prime = True
    # This for loop iterates over the range of potential factors
    for k in range(2, n):
        # This condition checks if k is a factor of n
        if n % k == 0:
            # If n % k == 0, then n is not prime
            print(f"{n} = {k} * {n // k}")
            is_prime = False
            break
    # This condition checks if is_prime is True
    if is_prime:
        # If n is prime, then print as output 
        # and append to list of prime numbers
        print(f"{n} is a prime number")
        prime_numbers_list.append(n)
        
print(f"\nList of prime numbers in range: {prime_numbers_list}")
```

The purpose of the code above is to check if the number it is prime, thus having only itself and 1 as a factor, it counts up from one and displays the factors if not prime, counting up from 2 to 11, then lists out the prime numbers between the set domain.
